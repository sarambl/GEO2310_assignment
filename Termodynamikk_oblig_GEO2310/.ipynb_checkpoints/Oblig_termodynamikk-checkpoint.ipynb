````python
class Moist_profile(Temperature_profile):
    def __init__(self, nr_boxes, top_pressure, temperatures, RH_profile, p0=1000., scale_H=8000):
        super().__init__(nr_boxes, top_pressure, temperatures=temperatures, p0=p0, scale_H=scale_H)
        # Set initiel mixing ratio
        self.w = #mixing ratio
        self.w_l =  #liquid mixing ratio, start with zero
        self.w_l_t0 = 0*self.w #liquid mixing ratio, start with zero
        self.w_s = # saturation mixing ratio
        self.relative_humidity= RH_profile

        return
    
    def update(self):
        self.w_s = self.sat_mix_ratio(self.sat_vap_pres( self.temperatures), self.p_mid)
        self.relative_humidity = self.w/self.w_s*100.
        return
    
    def sat_mix_ratio(self, sat_vap_pres,pressure):
        # use eq. 3.63 Wallace and Hobbs
        return 
    
    def partial_vap_pressure(self, w, pressure): #partial vapor pressure
        #use 3.59 Wallace and Hobbs
        epsilon=0.622
        return w/(w+epsilon)*pressure
    
        
    def sat_vap_pres(self, temp): #saturation vapor pressure
        # approximation used in Metpy with ref to https://journals.ametsoc.org/doi/abs/10.1175/1520-0493%281980%29108%3C1046%3ATCOEPT%3E2.0.CO%3B2
        sat_vap_pres_0C = 6.112 #* hPa
        T_celcius = temp - 273.15 # K--> Celcius
        sat_vap_pres = sat_vap_pres_0C * np.exp((17.67*T_celcius)/(T_celcius + 243.5))
        
        return sat_vap_pres
    
    def w2RH(self, w, pressure, temp): #relative humidity
        # eq. 3.64 wallace and Hobbs
        return w/self.sat_mix_ratio(self.sat_vap_pres(temp), pressure)*100 #%
    
    def RH2w(self, RH, pressure, temp):
        # eq. 3.64 wallace and Hobbs
        return RH/100.*self.sat_mix_ratio(self.sat_vap_pres(temp),pressure)
    
    def lift_box_1_step(self,i):
        if i==0: return False # if top index, the box can't be lifted
        theta_i = self.theta[i] # potential temperature in box i
        w_i = self.w[i] # vapor mixing ratio in box i
        w_i_l = self.w_l[i] # liquid water mixing ratio in box i
        pres_lifted = self.p_mid[i-1] # Pressure of box above
        temp_lifted = self.theta2temp_val(theta_i, pres_lifted) # calculate temperature when lifted
        theta_lifted = theta_i # potential temperature is constant when lifted before heat release
        
        # Calculate saturation mixing ratio when air is lifted.
        sat_mix_lifted= self.sat_mix_ratio(self.sat_vap_pres(temp_lifted),pres_lifted) 
        # If vapor mixing ratio in box higher than saturation mixing ratio, vapor condenses:
        if (sat_mix_lifted < w_i): 
            # vapor to condense:
            w_cond = w_i - sat_mix_lifted # difference between mixing ratio and sat mix ratio
            w_i_l = w_i_l + w_cond # Add difference to liquid water
            w_i = w_i-w_cond # Subtract difference from vapor
            temp_lifted = temp_lifted + Lv/cp*w_cond # Increase temperature accordingly
            theta_lifted = self.temp2theta_val(temp_lifted, pres_lifted)# theta_lifted + Lv/cp*diff # ????????????????
        # If the temperature when lifted (and potentially added latent heat from condensation.)
        if temp_lifted > self.temperatures[i-1]: # If lifted temperature higher than temperature above:
            # shift box above one down:
            self.temperatures[i] = self.theta2temp_val(self.theta[i-1], self.p_mid[i]) # set temperature 
            self.theta[i] = self.theta[i-1] # potential temperature does not change
            self.w[i] = self.w[i-1] # mixing ratio does not change
            self.w_l[i] = self.w_l[i-1] # liq mixing ratio does not change
            
            # Check if any liquid will evaporate when box is pushed down
            sat_mix_down= self.sat_mix_ratio(self.sat_vap_pres(self.temperatures[i]),self.p_mid[i]) 
            if self.w_l[i]>0 and sat_mix_down> self.w[i]:
                diff= sat_mix_down - self.w[i]
                w_evap= min(self.w_l[i], diff) # if w_l smaller than difference, evaporate all liq.
                self.w_l[i] = self.w_l[i] - w_evap
                self.w[i] = self.w[i] + w_evap
                self.temperatures[i]= self.temperatures[i] - Lv/cp*w_evap
                self.theta[i]= self.temp2theta_val(self.temperatures[i], self.p_mid[i])
            # Set temperature for box above:
            self.temperatures[i-1] = temp_lifted 
            self.theta[i-1] = theta_lifted
            self.w[i-1] = w_i
            self.w_l[i-1] = w_i_l
            box_lifted=True
        else: # if the temperature of the box when lifted is not higher than the box above, don't do anything
            box_lifted=False
        return box_lifted
    
    def lift_box(self,i):
        # lift one box i until equlibrium 
        j=i
        box_lifted = self.lift_box_1_step(j)
        while box_lifted:
            # if box lifted, try next index:
            j=j-1
            box_lifted = self.lift_box_1_step(j)
        if j!=i: # if j not equal i, the box has been lifted.
            box_lifted=True
        return box_lifted
            
            

    def equilibrium(self):
        # Lift unstable boxes until stable atmosphere
        i = len(self.p_mid)-1  # bottom layer index 
        while i>0: # start at bottom (top index) and work towards top.
            # Lift box i
            box_lifted= self.lift_box(i)
            # If box i is lifted, a new box is on index i
            # If not, we move one step up, i.e. i is reduced with one
            if not box_lifted:
                i=i-1
        self.update()
        return
  
            
    def theta2temp_val(self,theta, pres):
        return theta*(pres/self.p0)**self.r2cp
    def temp2theta_val(self,temp, pres):
        theta=temp*(self.p0/pres)**self.r2cp
        return theta

            
     
            
````
